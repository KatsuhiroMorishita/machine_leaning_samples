```python
# purpose: kerasCNN
# author: Katsuhiro MORISHITA
# memo: 
# created: 2018-02-17
import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.layers import Conv2D, MaxPooling2D
from keras.utils import np_utils
from matplotlib import pylab as plt
from PIL import Image
import numpy as np
import os


data_format = "channels_last"


def openfile(dir_name, data_format="channels_last"):
    """ 
    """
    image_list = []
    files = os.listdir(dir_name)    # 
    print(files)

    for file in files:
        root, ext = os.path.splitext(file)  # 
        if ext != ".jpg":
            continue

        path = os.path.join(dir_name, file) # 
        # 32x32pixel1[R,G,B]3
        # [R,G,B]0-255
        image = np.array(Image.open(path).resize((32, 32)))
        if data_format == "channels_first":
            image = image.transpose(2, 0, 1)   # [[Red],[Green],[Blue]] 
        image = image / 255.                   # 0-1
        image_list.append(image)        # image_list  
    
    return image_list

# 
img1 = openfile('1_train')
img2 = openfile('2_train')
x = np.array(img1 + img2)  # 
y = np.array([0] * len(img1) + [1] * len(img2))  # 
y = np_utils.to_categorical(y)                   # one-hot-encoding
print(x.shape)
print(y)
#exit()


# 
model = Sequential()
model.add(Conv2D(32, (3, 3), padding="same", data_format=data_format, input_shape=x.shape[1:]))  # 32, (3,3), input_shape1https://keras.io/ja/layers/convolutional/#conv2d
model.add(Activation('relu'))
model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))
model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(2))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(2))                # 2
model.add(Activation('sigmoid'))
model.add(Activation('softmax'))
opt = keras.optimizers.Adam(lr=0.0005, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0003) # lr
model.compile(optimizer=opt,       # 
      loss='categorical_crossentropy',
      metrics=['accuracy'])
print(model.summary())


# 
epochs = 100 # 1
batch_size = 8              # 
history = model.fit(x, y, 
    epochs=epochs, 
    batch_size=batch_size, 
    verbose=1, 
    validation_split=0.1,
    #validation_data=(x_test, y_test), # validation_datavalidation_split
    shuffle=True,           # 
    ) # lossaccmetrics


def plot_history(history):
    """ 
    from http://www.procrasist.com/entry/2017/01/07/154441
    """
    plt.plot(history.history['loss'],"o-",label="loss",)
    plt.plot(history.history['val_loss'],"^-",label="val_loss")
    plt.title('model loss')
    plt.xlabel('epoch')
    plt.ylabel('loss')
    plt.legend(loc='lower right')
    plt.grid()
    plt.yscale("log") # 
    plt.show()


# 
print(model.summary())                    # ()
open("model", "w").write(model.to_json()) # 
model.save_weights('param.hdf5')          # 
plot_history(history)                     # loss
```
